## Exercise 6.5 ISBN 

The International Standard Book Number (ISBN) is a numeric commercial book identifier which is intended to be unique. Publishers purchase ISBNs from an affiliate of the International ISBN Agency.

Write a code that 
- takes in a string input,
- return `True` if the string is either a valid ISBN-10 or ISBN-13 numbers.

Each line in the text file, `ISBN_EXERCISE.TXT` under the folder `resources`, contain ISBN-10 or ISBN-13 which can be valid or invalid.

Write a program to:
- print out the valid ISBNs in the file.
- print out the number of valid ISBNs in the file.

````python
# Rules for valid ISBN numbers:
# ISBN-10: 10 digits, 0-9, sum of position * digits mod 11 == 0 
# ISBN-13: 13 digits, sum of alternating 1 and 3 * digits mod 10 == 0

def check_wonky(isbn, num):
    digits = []
    for i in isbn:
        try:
            digits.append(int(i))
        except:
            continue
    if num == 10:
        if len(digits) != 10:
            return False
        total = sum((i + 1) * digits[i] for i in range(10))
        if total % 11 == 0:
            return True
    elif num == 13:
        if len(digits) != 13:
            return False
        total = sum(d * (1 if i % 2 == 0 else 3) for i, d in enumerate(digits))
        if total % 10 == 0:
            return True
    return False

valid = []

with open('./resources/ISBN_EXERCISE.txt', 'r') as f:
    data = f.readlines()
    for record in data:
        stuff = record.split(':')
        entry = stuff[0].strip()
        isbn = stuff[1].strip().replace('-','')
        if entry == 'ISBN-10' and check_wonky(isbn, 10):
            valid.append(record.strip())
        elif entry == 'ISBN-13' and check_wonky(isbn, 13):
            valid.append(record.strip())

print("Valid ISBNs:") 
for data in valid:
    print(data)
print(f"Number of Valid ISBNs: {len(valid)}")
````

A pixel holds the information of a color and the colors are often represented in a string starting with a hex `#` character followed by 3 pairs of hexadecimal digits. We will call this the hexadecimal representation of the pixel. For example, the color white is `#FFFFF` and the color silver is `#C0C0C0`.

### Task 1
Write a program code with the following specification:
- input a hexadecimal number as a string
- validate the input
- calculate the denary value of the hexadecimal number input
- output the denary value.

````python
# Format is #xx'yy'zz' where v' is the remainder 
# If remainder >= 2 digits, convert to alphabet 10 = A, 11 = B, 12 = C, 13 = D, 14 = E, 15 = F
# Denary is the reverse 
dicti = {
    'A' : 10,
    'B' : 11,
    'C' : 12,
    'D' : 13,
    'E' : 14,
    'F' : 15
}
ans = input("Enter a hexadecimal number as a string")
char = [character.upper() for character in ans]
try:
    total = 0
    for i in range(len(char)):
        if i == 0 and char[i].isdigit():
            total += (int(char[i]) * 16**(len(char)-1))
        elif char[i].isdigit() and i != 0:
            total += int(char[i]) * (16**(len(char)-i-1))
        else:
            total += dicti[char[i]] * (16**(len(char)-i-1))
    print(total)
except:
    print("Invalid input.")

````

Each pair of the hexadecimal digits actually represents the intensity of 3 primary colors, red, green and blue, that make up the color in the pixel. As such, instead of having the hexadecimal representation of the pixel, each pixel can be represented by a list of 3 denary values that represents the intensity of the colours, red, green and blue respectively, e.g., the color `#FFCO80` can be represented as the list `[255,192,128]`. We will call this representation of the pixel the `RGB representation`.

### Task 2

````python
import csv

def convertyconverty(ans):
    dicti = {
    'A' : 10,
    'B' : 11,
    'C' : 12,
    'D' : 13,
    'E' : 14,
    'F' : 15
    }
    char = [character.upper() for character in ans]
    try:
        total = 0
        for i in range(len(char)):
            if i == 0 and char[i].isdigit():
                total += (int(char[i]) * 16**(len(char)-1))
            elif char[i].isdigit() and i != 0:
                total += int(char[i]) * (16**(len(char)-i-1))
            else:
                total += dicti[char[i]] * (16**(len(char)-i-1))
            return total
    except:
        return "Invalid hexa :("

with open('./resources/TASK3_2.csv') as f:
    data = [row for row in csv.reader(f)]
    colors = []
    for row in data:
        row_colors = []
        for record in row:
            rgb = []
            record = record.strip('#')
            for i in range(0,6,2):
                hexed = record[i:i+2]
                rgb.append((convertyconverty(hexed)))
            row_colors.append(rgb)
        colors.append(row_colors)
    print(data)
    print(colors)

with open('./resources/MY_RGB_IMAGE.TXT', 'w') as f:
   writer = csv.writer(f)
   for row in colors:
        str_row = [f"({r},{g},{b})" for r,g,b in row]
        writer.writerow(str_row)
````

### Task 3
A 16 by 13 pixels image is stored its RGB representation in the text file `TASK3_3.CSV` in the `resources` folder, where each line represents a row in the image.
Write an additional program code with the following specification: 
- read the rgb values from the file `TASK3_3.CSV` and store them in a suitable format,
- write the new presentation into the file `MY_HEX_IMAGE.TXT`.

````python
def convertyconverto(rgb):
    dicti = {
        10: 'A',
        11: 'B',
        12: 'C',
        13: 'D', 
        14: 'E', 
        15: 'F'
    }
    hex = "#"
    for p in rgb:
        hexy = ""
        if (p // 16) < 10:
            hexy += str(p // 16)
        elif (p // 16) >= 10:
            hexy += dicti[p//16]
        if (p % 16) >= 10:
            hexy += dicti[p%16]
        if (p % 16 ) < 10:
            hexy += str(p%16)
        hex += hexy
    return hex

            
with open('./resources/TASK3_3.csv') as f:
    data = [row for row in csv.reader(f)]
    hexes = []
    for row in data:
        row_hexes = []
        for record in row:
            try:
                values = [int(i) for i in record.strip('[]').split(',')]
                row_hexes.append(convertyconverto(values))
            except Exception as e:
                row_hexes.append(f"error :( it's here: {e}")
        hexes.append(row_hexes)

with open('./resources/MY_HEX-IMAGE.TXT', 'w') as f:
    for row in hexes:
        hex_row = ' '.join(row)
        f.write(hex_row + '\n') 
````
